```python
def create_submat (match, mismatch, alphabet):
    sm = {}
    for c1 in alphabet:
        for c2 in alphabet:
            if (c1 == c2): 
                sm[c1+c2] = match
            else:
                sm[c1+c2] = mismatch
    return sm

#function to show how the score of an alignment, represented by a list with two sequences,containing gaps
def score_pos (c1, c2, sm, g): #
    if c1 == "−" or c2=="−": 
        return g
    else:
        return sm[c1+c2]

#function to build the S and T matrices, returning the maximun score
def smith_Waterman (seq1, seq2, sm, g): 
    S = [[0]]
    T = [[0]]
    maxscore = 0
    for j in range(1, len(seq2)+1):
 
        S[0].append(0) #adds an item to the end of the list.
        T[0].append(0) 
    for i in range(1, len(seq1)+1):
        S.append([0])
        T.append([0])
    for i in range(0, len(seq1)):
        for j in range(len(seq2)):
            s1 = S[i][j] + score_pos (seq1[i], seq2[j], sm, g); 
            s2 = S[i][j+1] + g
            s3 = S[i+1][j] + g
            b = max(s1, s2, s3)
            if b <= 0: 
                S[i+1].append(0) 
                T[i+1].append(0)
            else:
                S[i+1].append(b) 
                T[i+1].append(max3t(s1, s2, s3))
                if b > maxscore:
                    maxscore = b
    return (S, T, maxscore)
```
